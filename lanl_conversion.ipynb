```python
import requests
from tqdm import tqdm

url = "https://csr.lanl.gov/data-fence/1762740213/MlriKgZZwd-4muud6LZtFB4v2TM=/auth/lanl-auth-dataset-1-00.bz2"
output_file = "lanl_auth_dataset_1_00.bz2"

# Stream download with progress bar
response = requests.get(url, stream=True)
total_size = int(response.headers.get('content-length', 0))
block_size = 1024  # 1 KB

print(f"Downloading: {output_file}")
with open(output_file, "wb") as file, tqdm(
    total=total_size, unit='B', unit_scale=True, desc=output_file
) as progress_bar:
    for data in response.iter_content(block_size):
        file.write(data)
        progress_bar.update(len(data))

print("✅ Download complete!")
```

```python
import bz2, os

# Verify file size
print(f"File size: {os.path.getsize(output_file)/1e6:.2f} MB")

# Extract the .bz2 file
extracted_file = output_file.replace(".bz2", "")
with bz2.open(output_file, "rb") as f_in, open(extracted_file, "wb") as f_out:
    f_out.write(f_in.read())

print(f"Extracted: {extracted_file}")
```
